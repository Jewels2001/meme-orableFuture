```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("WorldSustainabilityDataset.csv")
```

```python
df.head()
```

```python
# data to use from world sustainability dataset
#  - renewable electricity output (%)
#  - renewable enery consumption (%)
#  - annual production based emissions of CO2


# other possible useful data
#  - population
#  - gini index
#  - total natural resources (% of GDP)
```

```python
df_use=df[["Country Name","Country Code","Year","Renewable electricity output (% of total electricity output) - EG.ELC.RNEW.ZS","Renewable energy consumption (% of total final energy consumption) - EG.FEC.RNEW.ZS","Annual production-based emissions of carbon dioxide (CO2), measured in million tonnes","Population, total - SP.POP.TOTL","Total natural resources rents (% of GDP) - NY.GDP.TOTL.RT.ZS"]]
```

```python
df_use
```

```python
df_use=df_use.rename(columns={"Renewable electricity output (% of total electricity output) - EG.ELC.RNEW.ZS":"Renewable_electricity_output","Renewable energy consumption (% of total final energy consumption) - EG.FEC.RNEW.ZS":"Renewable energy consumption","Annual production-based emissions of carbon dioxide (CO2), measured in million tonnes":"CO2_Emission","Population, total - SP.POP.TOTL":"Population","Total natural resources rents (% of GDP) - NY.GDP.TOTL.RT.ZS":"Natural Resource_Rents"})
```

```python
df_use=df_use.fillna(0)
```

```python
df_use
```

```python
import matplotlib.pyplot as plt
```

```python
countries = set(item for item in df_use["Country Name"])
```

```python
countries
```

```python
selected_Country=input("Select the single country")
```

```python
year=[]
Renewable_energy=[]
for i in range(0,len(df_use)):
  if df_use["Country Name"][i]== selected_Country:
      year.append(df_use["Year"][i])
      Renewable_energy.append(df_use["CO2_Emission"][i])
plt.plot(year,Renewable_energy , label=selected_Country)



plt.xlabel("Year")
plt.ylabel("Value")
plt.title("Line Graph for Countries")
# plt.legend()

# Show the plot
plt.show()
```

```python
from scipy.optimize import curve_fit
def linear_func(x,a,b):
  return [i*a+b for i in x]


popt,pcov=curve_fit(linear_func,year,Renewable_energy)
plt.plot(year,linear_func(year,*popt), label=selected_Country)



plt.xlabel("Year")
plt.ylabel("Value")
plt.title("Line Graph for Countries")
# plt.legend()

# Show the plot
plt.show()
```

```python
predict=np.array(range(2000,2100,1))

popt,pcov=curve_fit(linear_func,year,Renewable_energy)
plt.plot(predict,linear_func(predict,*popt), label=selected_Country)
print("Values in the following year")
print("2023:",linear_func([2023],*popt))
print("2028(5 years from now): ",linear_func([2028],*popt))
print("2033(10 years from now): ",linear_func([2033],*popt))
print("2048(25 years from now): ",linear_func([2048],*popt))
print("2073(50 years from now): ",linear_func([2073],*popt))
```
